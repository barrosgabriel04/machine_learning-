```python
import pandas as pd
```

```python
# carregando os dados do primeiro DataFrame e exibindo os 5 primerios registros  
# loading data from the first DataFrame and displaying the first 5 records
draft = pd.read_csv('draft.csv')
draft.head()
```

```python
# carregando os dados do segundo DataFrame e exibindo os 5 primerios registros
# loading data from the second DataFrame and displaying the first 5 records
combine = pd.read_csv('combine.csv')
combine.head()
```

```python
# exibindos registros duplicados
# displaying duplicate lines
draft[draft.duplicated()]
```

```python
# retirando do dataframe os registros duplicados
# removing duplicate records from the dataframe
draft = draft.drop_duplicates()
```

```python
draft[draft.duplicated()]
```

```python
# Exibindo as colunas do Dataframe de jogadores Draftados 
# Displaying Drafted players Dataframe columns
colunas_draft = draft.columns
colunas_draft
```

```python
# filtrando apenas os dados com registros a partir de 1987 e as colunas desejadas
# filtering only data with records from 1987 onwards and the desired columns
draft = draft[draft['draft'] > 1986]
draft = draft[['playerId','draft','pick','position','nameFirst','nameLast']]
```

```python
draft
```

```python
# exibido as colunas dos jogadores qeu participaram do combine
# displayed the columns of the players who participated in the combine
colunas_combine = combine.columns
colunas_combine
```

```python
# flitrando somente as colunnas desejadas 
# filtering only the desired columns
combine = combine[['playerId', 'combineYear', 'combinePosition',
       'combineHeight', 'combineWeight', 'nameFirst',
       'nameLast', 'position',
       'heightInches', 'weight', 'combine40yd', 'combineVert',
       'combineBench', 'combineShuttle', 'combineBroad']]
```

```python
# concatenando os dataframes e formando um so
# concatenating the dataframes and forming a single
geral = pd.concat([draft,combine])
```

```python
geral.head(100)
```

```python
# relacionando os dataframes atribuindo dados relacionados do mesmo jogador
# relating dataframes by assigning related data from the same player
geral = geral.groupby('playerId').sum()
```
